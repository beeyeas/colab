```python
# prompt: Identify a object in video using Google vision api

!pip install google-cloud-videointelligence

from google.cloud import videointelligence_v1p3beta1 as videointelligence

def analyze_video(video_uri):
    """Detects objects in a video."""

    client = videointelligence.VideoIntelligenceServiceClient()
    features = [videointelligence.Feature.OBJECT_TRACKING]

    operation = client.annotate_video(
        request={
            "input_uri": video_uri,
            "features": features,
        }
    )

    print("\nProcessing video for object annotations:")

    result = operation.result(timeout=90)

    print("\nFinished processing.")

    # Process object annotations
    for annotation in result.annotation_results:
        for object_annotation in annotation.object_annotations:
            print(f"Entity description: {object_annotation.entity.description}")
            for frame in object_annotation.frames:
                print(f"  Time offset: {frame.time_offset.seconds}.{frame.time_offset.nanos}")
                print(f"  Normalized bounding box:")
                print(f"    Left: {frame.normalized_bounding_box.left}")
                print(f"    Top: {frame.normalized_bounding_box.top}")
                print(f"    Right: {frame.normalized_bounding_box.right}")
                print(f"    Bottom: {frame.normalized_bounding_box.bottom}")

# Replace with your video URI
#video_uri = "gs://your-bucket/your-video.mp4"  # Example using Google Cloud Storage
video_uri = "path/to/your/local/video.mp4" # for local file

analyze_video(video_uri)
```
