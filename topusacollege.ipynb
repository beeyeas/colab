```python
# prompt: Generate code for maps with top 25 Rateed colleges in the USA, get the dataset from google. Give me python code  with marking overlay on the map.

# Install necessary libraries
!pip install plotly

import pandas as pd
import plotly.express as px

# Load the dataset (replace with the actual URL or local path)
# You'll need to find a dataset of top 25 US colleges with latitude and longitude.
# Example using a placeholder dataset:  https://raw.githubusercontent.com/beeyeas/colab/refs/heads/main/top_25_colleges_by_admission.csv
try:
    df = pd.read_csv('https://raw.githubusercontent.com/beeyeas/colab/refs/heads/main/top_25_colleges_by_admission.csv')

except:
    print("Error: Could not load the dataset. Please make sure the URL is correct and the file exists.")
    df = pd.DataFrame({
        'Name': ['College A', 'College B', 'College C'],
        'Rate': [1, 2, 3],
        'Latitude': [34.0522, 40.7128, 37.7749],
        'Longitude': [-118.2437, -74.0060, -122.4194]
    })

# Check if the necessary columns are present in the DataFrame
if not all(col in df.columns for col in ['Name', 'Rate', 'Latitude', 'Longitude']):
    print("Error: The dataset must have 'Name', 'Rate', 'Latitude', and 'Longitude' columns.")
    exit()

# Create the map
fig = px.scatter_mapbox(df,
                        lat="Latitude",
                        lon="Longitude",
                        hover_name="Name",
                        hover_data=["Rate"],
                        color="Rate",  # Color points by Rate
                        size=df['Rate'].str.rstrip('%').astype(float), # Remove '%' and convert to int
                        zoom=3,
                        height=600,
                        width=800,
                        mapbox_style="open-street-map")


fig.update_layout(title_text="Top Rated Colleges in the USA", title_x=0.5)

fig.show()
```
